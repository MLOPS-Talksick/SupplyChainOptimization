<h1> ARIMA Model Analysis </h1>

Auto-regressive Integrated Moving Average

(Univariate Time-Series Forcasing)

Pre-condition: The time-series data needs to be stationary (i.e. the variance of the log of the prices increases slower than that of a random walk / price-series that are found to be mean reverting)



```python
# Loading the data from Excel file
import sys
sys.path.append('..')

from dataReader import read_excel

data = read_excel('../data/commodity_demand_20190103_20241231.xlsx')

print(data.head())
```

<h3> Augmented Dickey Fuller Test </h3>

ADF is used to determine if time-series is stationary. 

$H_0$: The time-series is non-stationary

If $p$-value is $<$ signifcance level, we reject $H_0$ and assume that the data is stationary.

```python
from statsmodels.tsa.stattools import adfuller
import pandas as pd

# Define a concise ADF test function that returns key results (instead of printing them)
def adf_test(series):
    """Runs Augmented Dickey-Fuller test and returns a dict of results."""
    series_clean = series.dropna().reset_index(drop=True)
    adf_stat, p_value, used_lag, n_obs, crit_vals, icbest = adfuller(series_clean)
    return {
        'ADF Statistic': adf_stat,
        'p-value': p_value
    }

# List commodities, and define a p-value threshold
commodities = ["Corn", "Wheat", "Soybeans", "Sugar", "Coffee", "Beef", "Milk", "Chocolate"]
p_threshold = 0.05  # Example threshold

# Collect results in a list of dictionaries
results = []
for commodity in commodities:
    if commodity in data.columns:
        test_result = adf_test(data[commodity])
        results.append({
            'Commodity': commodity,
            'ADF Statistic': test_result['ADF Statistic'],
            'p-value': test_result['p-value'],
            'AboveThreshold': test_result['p-value'] > p_threshold
        })
    else:
        # If column not found, record the commodity but leave ADF data as None
        results.append({
            'Commodity': commodity,
            'ADF Statistic': None,
            'p-value': None,
            'AboveThreshold': None
        })

# Convert results to a DataFrame for concise display
results_df = pd.DataFrame(results)

# Print or display the full table
print("=== ADF Test Results ===")
display(results_df)

# Optionally, show which commodities exceed the p-value threshold
above_threshold_df = results_df[results_df['AboveThreshold'] == True]
if not above_threshold_df.empty:
    print(f"Commodities with p-value above {p_threshold}:")
    display(above_threshold_df[['Commodity', 'p-value']])
    print("If p-value is above threshold, the series is non-stationary and we need to find the order of differencing.")
else:
    print(f"No commodities have a p-value above {p_threshold}.")
```

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

# How many times do you want to difference the series?
n_diffs = 2  # e.g., this checks D=0 (original), D=1, D=2

# These are the commodities deemed "non-stationary" in your previous step
non_stationary_commodities = above_threshold_df['Commodity'].dropna()

for commodity in non_stationary_commodities:
    series_clean = data[commodity].dropna()

    # Create a figure with (n_diffs + 1) subplots in a single row
    fig, axs = plt.subplots(1, n_diffs + 1, 
                            figsize=(5 * (n_diffs + 1), 4), 
                            sharey=False)

    # If there's only one subplot (edge case, n_diffs=0), axs won't be a list
    # so let's make it a list for consistent indexing
    if n_diffs == 0:
        axs = [axs]

    # Loop over each differencing order
    for i, diff_order in enumerate(range(n_diffs + 1)):
        if diff_order == 0:
            # No differencing: the original series
            data_to_plot = series_clean
            diff_label = "Original"
        else:
            # Difference the series diff_order times
            data_to_plot = series_clean.diff(diff_order).dropna()
            diff_label = f"Diff={diff_order}"
        
        ax = axs[i]
        plot_acf(data_to_plot, ax=ax, lags=40, 
                 title=f'{commodity}: {diff_label}')

    plt.tight_layout()
    plt.show()
```

```python
# I wasn't really satisified with ndiffs, so I will determine my own differencing order
# Here is ndiffs for context:
from pmdarima.arima.utils import ndiffs

results = []
for commodity in commodities:
    if commodity in data.columns:
        series_clean = data[commodity].dropna()
        
        # ndiffs returns how many differences are recommended (0 means already stationary)
        d_value = ndiffs(series_clean, test='adf')  # or 'kpss'/'pp'
        
        results.append({
            "Commodity": commodity,
            "Recommended Differences": d_value
        })
    else:
        # If column not found, record None (or 0) as you prefer
        results.append({
            "Commodity": commodity,
            "Recommended Differences": None
        })

# Convert to a concise DataFrame
results_df = pd.DataFrame(results)

# Show the recommended differencing for each commodity
print("=== ndiffs Results (ADF) ===")
display(results_df)

# Even when we said earlier that the p-value is above the threshold, it shows that the recommended differencing is 0
# For sugar I will try differencing with 2 and for coffee I will try differencing with 1
```

<h3> p </h3>

The order of the auto-regressive term. (i.e. the number of lags to be used as predictors)

We can inspect the Partial Autocorrelation (PACF) plot which represents the correlation between the series and its lags.

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf

# 1) Specify differencing orders for each commodity
diff_orders = {
    "Corn": 1,
    "Wheat": 1,
    "Soybeans": 0,
    "Sugar": 2,
    "Coffee": 1,
    "Beef": 0,
    "Milk": 1,
    "Chocolate": 0
}

def plot_diff_and_pacf_custom(data, diff_dict, pacf_lags=40):
    """
    For each commodity in 'diff_dict', difference the series by the specified amount
    and plot:
      1) The differenced series
      2) The PACF of that differenced series
    If a commodity isn't in 'diff_dict', defaults to 0 differences.
    """
    for commodity, diff_order in diff_dict.items():
        # Check if the column exists
        if commodity not in data.columns:
            print(f"Column '{commodity}' not found in DataFrame.")
            continue
        
        # Get the time series and drop NaNs
        series = data[commodity].dropna()
        
        # Apply differencing 'diff_order' times, or 0 if not specified
        if diff_order > 0:
            diff_series = series.diff(diff_order).dropna()
        else:
            diff_series = series  # no differencing
        
        # Create figure: left subplot for differenced series, right for PACF
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))
        
        # Plot the differenced series
        ax1.plot(diff_series)
        ax1.set_title(f"{commodity}: Differenced {diff_order} time(s)")
        
        # Plot PACF
        plot_pacf(diff_series, ax=ax2, lags=pacf_lags, title=f"PACF: {commodity}")
        
        plt.tight_layout()
        plt.show()

plot_diff_and_pacf_custom(data, diff_orders, pacf_lags=50)

# PACF lag 7 seems to have some sort of significance for each of the commodities (since the significance seems to spike in multiples of 7 after the initial spikes)
```
